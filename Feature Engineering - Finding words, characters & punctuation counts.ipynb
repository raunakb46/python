```python
import pandas as pd
import numpy as np
import string
```

```python
df = pd.read_csv('amazon_alexa.tsv', delimiter = '\t')
```

```python
df.shape
```

```python
df.head()
```

```python
# Calculating character count in review

df['char_count'] = df['verified_reviews'].apply(len)
df.head()
```

```python
# Calculating word count

df['word_count'] = df['verified_reviews'].apply(lambda x: len(x.split()))
df.head()
```

```python
# Calculating the word density

df['word_density'] = df['char_count'] / (df['word_count'] + 1)
df.head()
```

```python
# Importing list of punctuations

punc = string.punctuation

# Calculating punctuation count
df['punc_count'] = df['verified_reviews'].apply(lambda x: len("".join(_ for _ in x if _ in punc)))
df.head()
```

```python
# Summarizing the dataframe

df[['char_count', 'word_count', 'word_density', 'punc_count']].describe()
```
