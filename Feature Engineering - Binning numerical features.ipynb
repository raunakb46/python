```python
df['Age'].value_counts()
```

```python
# Categorizing or bining 'Age'

def age_bin(x):
    if x > 20 and x < 30:
        return '20 - 30'
    elif x > 30 and x < 40:
        return '30 - 40'
    elif x > 40 and x < 50:
        return '40 - 50'
    else:
        return '50 - 60'
```

```python
df['Age'] = df['Age'].apply(age_bin)
df['Age'].value_counts()
```

```python
df.head()
```

```python
df['DistanceFromHome'].value_counts()
```

```python
# Categorizing the distance as 'Near', 'Far' and 'Extremely Far'

def dist(x):
    if x <= 2:
        return 'Near'
    elif x > 2 and x <= 5:
        return 'Far'
    elif x > 5:
        return 'Extremely far'
```

```python
df['DistanceFromHome'] = df['DistanceFromHome'].apply(dist)
df['DistanceFromHome'].value_counts()
```

```python
# Bining experience on the basis of years of experience

def exp(y):
    if y <= 2:
        return 'Novice'
    elif y > 2 and y <= 5:
        return 'Intermediate'
    elif y > 5:
        return 'Experienced'
```

```python
df['TotalWorkingYears'] = df['TotalWorkingYears'].apply(exp)
df['TotalWorkingYears'].value_counts()
```
