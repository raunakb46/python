```python
# Checking for distinct values in 'EmployeeCount'

df['EmployeeCount'].value_counts()
```

All the values are same i.e. 1

```python
# Checking for distinct values in 'Over18'

df['Over18'].value_counts()
```

All the values are same i.e Y

```python
# Checking for distinct values in 'StandardHours'

df['StandardHours'].value_counts()
```

All the values are same i.e. 80

```python
# 'EmployeeCount', 'Over18' and 'StandardHours' has same values throughout the column and no variation can be seen.
# This won't affect the target variable. Hence, dropping the columns.

df = df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis = 1)
```

```python
# Checking for number of unique values in 'EmployeeNumber'

df['EmployeeNumber'].nunique()
```

```python
# 'EmployeeNumber' has all unique values across the column and that means high variation.
# This again will not affect the target variable. Hence, dropping it.

df = df.drop(['EmployeeNumber'], axis = 1)
```

```python
df.columns
```

```python
# Finding correaltion among all the variables

plt.rcParams['figure.figsize'] = (18, 9)
sns.heatmap(df.corr(), annot = True)
```

```python
# 'JobLevel' is highly correlated with 'MonthlyIncome'
# 'YearsAtCompany' is highly correlated with 'YearsInCurrentRole' and 'YearsWithCurrManager'

# Dropping the highly correlated columns

df = df.drop(['JobLevel', 'YearsInCurrentRole', 'YearsWithCurrManager'], axis = 1)
```
