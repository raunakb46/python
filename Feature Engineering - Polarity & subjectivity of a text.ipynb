```python
import pandas as pd
import numpy as np
from textblob import TextBlob
```

```python
df = pd.read_csv('amazon_alexa.tsv', delimiter = '\t')
```

```python
df.shape
```

```python
df.head()
```

```python
# Getting the length of reviews

df['length'] = df['verified_reviews'].apply(len)
df.head()
```

```python
# Polarity tells about the sentiment of text i.e positive(+1), negative(-1) or neutral(0)
# Polarity lies between -1 to 1

def polartxt(txt):
    textblob = TextBlob(str(txt.encode('utf-8')))
    polar = textblob.sentiment.polarity
    return polar

df['polarity'] = df['verified_reviews'].apply(polartxt)
df.head()
```

```python
# Subjectivity tells if the text is subjective(1) or not(0) i.e briefly described
# Subjectivity lies between 0 and 1

def subjectivetxt(txt):
    textblob = TextBlob(str(txt.encode('utf-8')))
    sub = textblob.sentiment.subjectivity
    return sub

df['subjectivity'] = df['verified_reviews'].apply(subjectivetxt)
df.head()
```
