```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Student.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
# Adding 'math score' , 'reading score' and 'writing score' and storing them in new feature

df['total score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
df.head()
```

```python
# Creating a new feature - 'percentage'

df['percentage'] = (df['total score'] / 300) * 100
```

```python
df.head()
```

```python
# Function to calculate grade

def grade_cal(x):
    if x > 80:
        return 'A'
    elif x > 70 and x < 80:
        return 'B'
    elif x > 60 and x < 70:
        return 'c'
    elif x > 50 and x < 60:
        return 'D'
    elif x > 40 and x < 50:
        return 'E'
    else:
        return 'F'
```

```python
df['grade'] = df['percentage'].apply(grade_cal)
```
