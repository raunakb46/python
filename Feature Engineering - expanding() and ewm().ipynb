```python
import pandas as pd
import numpy as np
```

```python
# Creating new dataframe

df = pd.DataFrame({'A':[0, 1, 2, np.nan, 4, 5, 6, np.nan, np.nan]})
df
```

```python
# Expanding the sum()
# The third no. is the sum of previous 2 nos. and sum is 'NaN' if no previous no. is present

df.expanding(2).sum()
```

```python
# Fifth no. is the sum of previous 4 nos.

df.expanding(4).sum()
```

```python
# Expanding the correlation function

df.expanding(3).corr()
```

```python
# Expanding kurt()

df.expanding(3).kurt()
```

```python
df.expanding(3).agg('mean')
```

```python
df.expanding(3).agg(['mean', 'median', 'max'])
```

```python
# ewm() --> Exponentially Weighted Window Function

df.ewm(com = 0.5).mean()
```

```python
df.ewm(span = 5).std()
```

```python
df.ewm(span = 5).corr()
```

```python
df.ewm(span = 5).var()
```
